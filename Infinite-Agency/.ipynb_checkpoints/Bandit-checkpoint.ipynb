```python
import os, sys
import glob
import pandas as pd
import numpy as np
import scipy.special
import operator

from collections import Counter, defaultdict
from time import time
import datetime

import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

from bokeh.charts import Histogram, defaults, vplot, hplot, show, output_file
```

```python
font = {'family' : 'monospace',
        'weight' : 'medium',
        'size'   : 24}

matplotlib.rc('font', **font)
```

```python
today = str(datetime.date.today())
```

```python
newpath = r'/home/rafa/github/For Businesses/Infinite-Agency/Results-From-{}/'.format(today)
```

```python
paths = []
with open('Results-From-{}/clean_paths-{}.txt'.format(today, today)) as f:
    for line in f:
        mini_path = []
        for item in line[1:-3].split(','):
            mini_path.append(item)
        paths.append(mini_path)
```

```python
path_dict = defaultdict(list)
for path in paths:
    for index, ad in enumerate(path):
        if index == len(path) - 1:
            path_dict[ad].append(1)
        else:
            path_dict[ad].append(0)
```

```python
counter_list = dict()
for key in path_dict.keys():
    cnt = Counter()
    for item in path_dict[key]:
        cnt[item] += 1
    counter_list[key] = cnt
```

```python
zero_list = []
one_list = []
key_list = []
for key in counter_list.keys():
    key_list.append(key)
    zero_list.append(counter_list[key][0])
    one_list.append(counter_list[key][1])
```

```python
xvalues = range(len(zero_list))
fig = plt.figure(figsize=(10, 10))
plt.grid(zorder = 0, linestyle = 'dashed', color = '#acaaa8')
p1 = plt.bar(xvalues, zero_list, ec = 'k', color='#A39D85', label = '0', zorder = 3)
p2 = plt.bar(xvalues, one_list, ec = 'k', color = '#FF8C00', bottom=zero_list, label = '1', zorder = 3)
plt.legend()
plt.title('Last Touch Attributions')
plt.ylabel('Number of Times Served')
plt.xticks(xvalues, 
           key_list, 
           rotation = 90)

plt.savefig(newpath + 'Last Touch Attributions-{}.png'.format(today))
plt.show()
```
