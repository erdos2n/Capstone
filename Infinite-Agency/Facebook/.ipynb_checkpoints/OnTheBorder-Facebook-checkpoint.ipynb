```python
import os, sys
import glob
import pandas as pd
import numpy as np
import scipy.special
import operator
from datetime import date

from collections import Counter

import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

from bokeh.charts import Histogram, defaults, vplot, hplot, show, output_file
```

```python
font = {'family' : 'monospace',
        'weight' : 'medium',
        'size'   : 18}

matplotlib.rc('font', **font)

today = pd.to_datetime(date.today())
```

```python
def drop_nonunique(big_frame):
    for column in big_frame.columns:
        if len(big_frame[column].unique()) == 1:
            big_frame.drop(column, axis=1, inplace=True)
    print big_frame.shape

def make_datelist(big_frame):
    datelist = []
    for column in big_frame.columns:
        if 'Date' in column:
            datelist.append(column)
    return datelist

def change_date(big_frame, datelist):
    datelist = []
    for column in datelist:
        big_frame[column] = pd.to_datetime(big_frame[column], infer_datetime_format=False)
    return big_frame
```

--------------------------------------

```python
df = pd.read_csv('Facebook Insights Data Export - Estrella Jalisco - 2017-05-12.csv')
df.shape
```

```python
df.dropna(axis = 1, how = 'all', inplace=True)
# drop_nonunique(big_frame)
df.fillna(value = 0, inplace=True)
df.drop(df.index[0], inplace=True)
df.shape
```

```python
for column in df.columns:
    if 'Array' in df[column].unique():
        df.drop(labels = column, axis = 1, inplace=True)
```

```python
df.shape
```

```python
drop_nonunique(df)
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True)
```

```python
df = df[df['Date'] < today]
```

```python
df_described = df.describe()
```

```python
df_described.to_csv('Full_Stats.csv')
```

# CREATE IMPRESSIONS DATAFRAME

```python
impression_list = []
for column in df.columns:
    if 'pression' in column:
        impression_list.append(column)
impression_list = impression_list[0:-1]
```

```python
impression_df = df[impression_list]
```

```python
for column in impression_df.columns:
    new_column = [int(column_object) for column_object in impression_df[column]]
    impression_df[column] = new_column
```

```python
impression_described = impression_df.describe()
format = lambda x: "{0:.2f}".format(x) #Alternatively, format = lambda x: '%.2f' % x
impression_described = impression_described.applymap(format)
impression_described.to_csv('Impression_Stats.csv')
```

```python
impression_described
```

```python
sm = pd.scatter_matrix(impression_df, figsize=(100, 100), alpha = 0.5, c = '#383632', grid=True, diagonal='kde')
[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]
[s.get_yaxis().set_label_coords(-0.75, 0.5) for s in sm.reshape(-1)]
plt.savefig(r"Impressions_Plots.png")
```
